# ES-LLMs: Pedagogical Orchestration Demo

This notebook provides a minimal executable demonstration of the Subsumption Architecture from the AIED 2026 paper: 
**From Untamed Black Box to Interpretable Pedagogical Orchestration**.

You will see how exactly the Orchestrator suppresses unsafe LLM generations (like giving away the answer) based on the student's attempt state.

```python
import random

class MockEthicsBot:
    def check_safety(self, student_action, attempt_count):
        if attempt_count == 0 and "help" in student_action.lower():
            return "DENY_HINT"
        return "SAFE"

class MockScaffoldBot:
    def generate_hint(self):
        return "HINT_FULL"

class SubsumptionOrchestrator:
    def __init__(self):
        self.ethics = MockEthicsBot()
        self.scaffold = MockScaffoldBot()

    def execute_policy(self, student_input, attempt_count):
        print(f"\nReceived Student Input: {student_input}")
        print(f"Current Attempt Count: {attempt_count}")
        
        # Layer 1: Safety (Highest Priority)
        safety_action = self.ethics.check_safety(student_input, attempt_count)
        if safety_action == "DENY_HINT":
            print("[ORCHESTRATOR] Subsumption Triggered! Safety suppresses Scaffolding.")
            return "Please attempt the problem before asking for a hint."
            
        # Layer 2: Pedagogy (Lower Priority)
        hint_action = self.scaffold.generate_hint()
        return f"Here is a hint: {hint_action}"

# Initialize the system
orchestrator = SubsumptionOrchestrator()

print("--- SCENARIO 1: Premature Help Request ---")
print("Tutor Action:", orchestrator.execute_policy("Idk, just give me the answer.", attempt_count=0))

print("\n--- SCENARIO 2: Valid Help Request ---")
print("Tutor Action:", orchestrator.execute_policy("I tried adding the numbers but got it wrong. Help?", attempt_count=1))
```
